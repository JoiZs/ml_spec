```python
from matplotlib import pyplot as plt
import numpy as np
```

### Sample data from np

House Pricing LN project

x - size of the house
y - price of the house
x_train = [ 1.0 2.0], y_train = [ 100 300]

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([100, 300])
```

Training Model for linear regress with Univariate - f(x) = wx + b

```python
def modelFnwb(x_train):
    w = 200
    b = -100
    m = x_train.shape[0]
    mdi = np.zeros(m)
    
    for i in range(m):
        mdi[i] = w*x_train[i] + b

    return mdi
    
```

```python
tmp_fwb = modelFnwb(x_train)

# Plot orginal data points
plt.scatter(x_train, y_train, marker="x", c="r", label="actual pt")

# Plot model points
plt.plot(x_train, tmp_fwb, c="b", label="prediction")

plt.xlabel("size")
plt.ylabel("price")

plt.legend()
plt.show()
```

```python
# Predict for 1.4 (1400 sq) size
prediction = modelFnwb(np.array([1.4]))
print(f"price of 1.4ksq house: {prediction[0]}k$")
```
